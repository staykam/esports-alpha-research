```python
# main.py
import nbformat
from nbconvert.preprocessors import ExecutePreprocessor

# Load the .ipynb file
with open("V2_bo3_scraper.ipynb", "r") as nb_file:
    nb_contents = nb_file.read()

# Execute the notebook (run all cells)
nb = nbformat.reads(nb_contents, as_version=4)
ep = ExecutePreprocessor(timeout=600, kernel_name="python3")
ep.preprocess(nb, {"metadata": {"path": "./"}})

# Extract the "bo1s" DataFrame
for cell in nb.cells:
    if cell.cell_type == "code" and "bo1s" in cell.source:
        exec(cell.source)
        break

# Now you can use the "bo1s" DataFrame in your current script
bo1s.head # Example: Print the first few rows of the DataFrame
```
