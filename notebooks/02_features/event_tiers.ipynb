### Code to get Tiers of events

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

#Set up the webdriver
driver = webdriver.Firefox()

#Define the list of tiers and their URLs
tiers = [
('S-tier', 'https://liquipedia.net/counterstrike/S-Tier_Tournaments'),
('A-tier', 'https://liquipedia.net/counterstrike/Major_Tournaments'),
('B-tier', 'https://liquipedia.net/counterstrike/B-Tier_Tournaments'),
('C-tier', 'https://liquipedia.net/counterstrike/Minor_Tournaments'),
('Qualifiers', "https://liquipedia.net/counterstrike/Qualifier_Tournaments"),
]

#Define a list to store the events and their tiers
events = []

#Iterate over the tiers
for tier, url in tiers:
    #Navigate to the tier webpage
    driver.get(url)
    # Find the grid table
    try:
        if tier == 'Qualifiers':
            gridtable = driver.find_element(By.CSS_SELECTOR, 'div.gridTable.tournamentCard.Qualifiers')
            # Find all the grid rows within the grid table
            gridrows = gridtable.find_elements(By.CSS_SELECTOR, 'div.gridRow')
            # Iterate over the grid rows
            for gridrow in gridrows:
                # Wait for the a tag within the grid row to be present
                tournament_link = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.gridCell.Tier.Header a[href*='-Tier']")))
                # Extract the title attribute value from the a tag
                event_tier = tournament_link.get_attribute('title')
                # Find the tournament name
                tournament_name_element = gridrow.find_element(By.CSS_SELECTOR, "div.gridCell.Tournament.Header a")
                event_name = tournament_name_element.get_attribute('title')
                # Add the event and its tier to the list
                events.append((event_name, event_tier))
        else:
            gridtable = driver.find_element(By.CSS_SELECTOR, 'div.gridTable.tournamentCard.Tierless')
            # Find all the grid rows within the grid table
            gridrows = gridtable.find_elements(By.CSS_SELECTOR, 'div.gridRow')
            # Iterate over the grid rows
            for gridrow in gridrows:
                # Find the a tag within the grid row
                tournament_link = gridrow.find_element(By.CSS_SELECTOR, "div.gridCell.Tournament.Header a")
                # Extract the title attribute value from the a tag
                event_name = tournament_link.get_attribute('title')
                # Add the event and its tier to the list
                events.append((event_name, tier))
    except Exception as n:
        print("Error occurred: ", n)

#Close the webdriver
driver.quit()
```

#### Solving different format of writing S-tier or S-Tier ,etc. Also making sure that the tier is always just the tier not extra words like "Tournaments"

```python
#Initalise a list to store all the rectifcations of the event names
modified_events = []

# Iterate through the events list
for event, tier in events:
    # Extract the tier value from the tier string
    if tier == "S-tier":
        tier_value = tier.lower()
    elif tier == "A-tier":
        tier_value = tier.lower()
    elif tier == "B-tier":
        tier_value = tier.lower()
    elif tier == "C-tier":
        tier_value = tier.lower()
    else:
        # If the tier string is not one of the expected values, extract the tier value from the string
        tier_value = tier.split(" ")[0].lower()

    # Create a tuple with the event name and modified tier value
    modified_event = (event, tier_value)

    # Add the modified event to the modified_events list
    modified_events.append(modified_event)

# Print the modified events list
#for event, tier in modified_events:
#    print(f"{event} ({tier})")
```

```python
import pandas as pd

# Define a function to extract the organizer, region, and tier from the dictionary
def extract_info(event_dict):
    event, tier = event_dict
    if "Series" in event:
        # Split the event name by "Series" and take the first part as the organizer
        organizer = event.split("Series")[0].strip()
        # Split the event name by "Series" and take the second part as the region
        region = event.split("Series")[1].strip()
    else:
        # If "Series" is not in the event name, use the entire event name as the organizer
        organizer = event.strip()
        # Set the region to None
        region = None
    # Return the organizer, region, and tier as a tuple
    return organizer, region, tier

# Create a list of tuples with the extracted information
info_list = [extract_info(event_dict) for event_dict in events]

# Create a new dataframe with the extracted information
df = pd.DataFrame(info_list, columns=["Organizer", "Region Played", "Tier"])

# Print the dataframe
df
```
