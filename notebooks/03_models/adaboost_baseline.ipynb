```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load the dataset
df = pd.read_excel(r'C:\Users\Coco\OneDrive - Erasmus University Rotterdam\Coding\Betting hltv\player_stats\general_df_test_stats.xlsx')

# Extract relevant features and target variable
X = df[['win_rate_team1', 'avg_rounds_won_team1', 'avg_rounds_lost_team1', 'win_rate_team2', 'avg_rounds_won_team2', 'avg_rounds_lost_team2']]
y = df['winning_team']

# Convert the target variable to binary format
y = y.apply(lambda x: 1 if x == 'Team 1' else 0)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create an instance of the AdaBoostClassifier
clf = AdaBoostClassifier(n_estimators=100, learning_rate=1.0)

# Train the model on the training data
clf.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = clf.predict(X_test)

# Evaluate the performance of the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print('Accuracy:', accuracy)
print('Precision:', precision)
print('Recall:', recall)
print('F1 score:', f1)
```
